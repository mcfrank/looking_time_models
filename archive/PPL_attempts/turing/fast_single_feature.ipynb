```julia
using DynamicHMC, Turing
using StatsPlots

```

```julia
@model function demo_model(cats, zs, n_sample)
  mu ~ Normal(0, 1)
  sigma ~ Gamma(1, 1)
  ys = Vector{Float64}(undef, 5)
  for i in 1:5
    ys[i] ~ Normal(mu, sigma)
  end
  for i in 1:n_sample
    zs[i] ~ Normal(ys[cats[i]], 0.5)
  end
  return mu, sigma
end
```

```julia
# Simulate data
n_sample = 50
ys = [0.8, 5.2, 1.2, 1.5, -0.4]
# ys = [10.8, 10.2, 11.2, 12.5, 10.6]
cat_dist = Categorical([0.1, 0.5, 0.2, 0.1, 0.1])
cats = rand(cat_dist, n_sample)

zs = [rand(Normal(ys[cats[i]], 0.5)) for i in 1:n_sample]

#  Run sampler, collect results
println(cats[1:5])
println(zs[1:5])
```

```julia
@time chn = sample(demo_model(cats, zs, n_sample), DynamicNUTS(), 20000);
```

```julia
# Plot and save results
p = plot(chn)
# savefig("demo-plot.pdf")
```
